```
#user defined functions
def display_menu():
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Update Task")
    print("4. Delete Task")
    print("5. Mark Task as Complete")
    print("6. Exit")
```

```
tasks = []

def add_task(task_name):
    task = {
        'name': task_name,
        'completed': False
    }
    tasks.append(task)

def view_tasks():
    for index, task in enumerate(tasks):
        status = "✔️" if task['completed'] else "✖️"
        print(f"{index + 1}. {task['name']} [{status}]")
```

```
#updating and deleting task
def update_task(index, new_name):
    if 0 <= index < len(tasks):
        tasks[index]['name'] = new_name

def delete_task(index):
    if 0 <= index < len(tasks):
        tasks.pop(index)
```

```
#making task complete
def mark_task_complete(index):
    if 0 <= index < len(tasks):
        tasks[index]['completed'] = True
```

```
#main loop
while True:
    display_menu()
    choice = input("Select an option: ")

    if choice == '1':
        task_name = input("Enter task name: ")
        add_task(task_name)
    elif choice == '2':
        view_tasks()
    elif choice == '3':
        index = int(input("Enter task number to update: ")) - 1
        new_name = input("Enter new task name: ")
        update_task(index, new_name)
    elif choice == '4':
        index = int(input("Enter task number to delete: ")) - 1
        delete_task(index)
    elif choice == '5':
        index = int(input("Enter task number to mark as complete: ")) - 1
        mark_task_complete(index)
    elif choice == '6':
        break
    else:
        print("Invalid choice. Please try again.")

```
